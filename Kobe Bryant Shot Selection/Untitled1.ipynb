```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.countplot(y_train)
```

```python
sns.pairplot(X_train, vars=['loc_x', 'loc_y', 'lat', 'lon',  'shot_distance']
                                                 , hue='shot_made_flag', size=3)
```

```python
f, axarr = plt.subplots(8, figsize=(15, 25))

sns.countplot(x="combined_shot_type", hue="shot_made_flag", data=X_train, ax=axarr[0])
sns.countplot(x="season", hue="shot_made_flag", data=X_train, ax=axarr[1])
sns.countplot(x="period", hue="shot_made_flag", data=X_train,  ax=axarr[2])
sns.countplot(x="playoffs", hue="shot_made_flag", data=X_train, ax=axarr[3])
sns.countplot(x="shot_type", hue="shot_made_flag", data=X_train, ax=axarr[4])
sns.countplot(x="shot_zone_area", hue="shot_made_flag", data=X_train, ax=axarr[5])
sns.countplot(x="shot_zone_basic", hue="shot_made_flag", data=X_train, ax=axarr[6])
sns.countplot(x="shot_zone_range", hue="shot_made_flag", data=X_train, ax=axarr[7])

axarr[0].set_title('Combined shot type')
axarr[1].set_title('Season')
axarr[2].set_title('Period')
axarr[3].set_title('Playoffs')
axarr[4].set_title('Shot Type')
axarr[5].set_title('Shot Zone Area')
axarr[6].set_title('Shot Zone Basic')
axarr[7].set_title('Shot Zone Range')

plt.tight_layout()
```
