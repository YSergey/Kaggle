```python
import os
import pandas as pd 

base_dir = '/Volumes/KESU/BNP_Paribas_Cardif_Claims_Management'
dataset = os.path.join(base_dir, './dataset')
```

```python
df_train = pd.read_csv(os.path.join(dataset, 'train.csv'))
df_test = pd.read_csv(os.path.join(dataset, 'test.csv'))
```

```python
print('Train data')
print([col for col in df_train.columns if df_train[col].isnull().sum() > 0])
print('Test data')
print([col for col in df_test.columns if df_test[col].isnull().sum() > 0])
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical_vars = list(df_train.select_dtypes(include=numerics).columns)
df_train = df_train[numerical_vars]
df_train.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_train.drop(labels=['ID', 'target'], axis=1), df_train['target'], 
                                                                                        test_size=.3, random_state=42)
X_train.shape, X_test.shape
```

```python
mi = mutual_info_classif(X_train.fillna(0), y_train, random_state=42)
mi 
```

```python
mi = pd.Series(mi)
mi.index =X_train.columns
mi.sort_values(ascending=True)
```
