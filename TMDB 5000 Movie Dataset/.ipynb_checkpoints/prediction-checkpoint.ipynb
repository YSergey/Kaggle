```python
import pandas as pd

df_movies = pd.read_csv('tmdb_5000_movies.csv')
df_credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
df_movies.head(3)
```

```python
df_credits.head(3)
```

```python
import json

json_col = ['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages']

for col in json_col:
    df_movies[col]=df_movies[col].apply(json.loads)
    arr = []
    
    for i in range(len(df_movies[col])):
        l_name = [d.get('name') for d in df_movies[col][i]]
        l_name = str(l_name)
        arr.append(l_name)
    
    df_movies[col] = pd.DataFrame(arr)
```

```python
df_movies.head()
```

```python
df_movies['genres']=df_movies['genres'].str.strip('[]').str.replace(' ','').str.replace("'",'')
df_movies['genres']=df_movies['genres'].str.split(',')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplots(figsize=(12,10))
list1=[]
for i in df_movies['genres']:
    list1.extend(i)
    
ax=pd.Series(list1).value_counts()[:10].sort_values(ascending=True).plot.barh(width=0.9,color=sns.color_palette('summer_r',10))

for i, v in enumerate(pd.Series(list1).value_counts()[:10].sort_values(ascending=True).values): 
    ax.text(.8, i, v,fontsize=12,color='white',weight='bold')
ax.patches[9].set_facecolor('r')
plt.title('Top Genres')
plt.show()
```
