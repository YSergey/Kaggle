```python
import pandas as pd

X_train = pd.read_csv('train.csv')

y_train = X_train['Cover_Type']
X_train = X_train.drop(['Cover_Type'], axis=1)
X_train.head()
```

```python
X_test = pd.read_csv('test.csv')
X_test.head()
```

```python
X_train.isnull().sum()
```

```python
X_test.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

pip_log = Pipeline([('scl', StandardScaler()), ('est', LogisticRegression(random_state=1))])
pip_gbc = Pipeline([('scl', StandardScaler()), ('est', GradientBoostingClassifier(random_state=1))])

# fit & evaluation
pipe_names = ['Logistic','GradientBoosting']
pipe_lines = [pip_log, pip_gbc]
for (i,pipe) in enumerate(pipe_lines):
    pipe.fit(X_train, y_train.as_matrix().ravel())
    #print('%s: %.3f'%(pipe_names[i], accuracy_score(y_test.as_matrix().ravel(), pipe.predict(X_test))))
    pipe.predict(X_test)
    print('OK')
```
